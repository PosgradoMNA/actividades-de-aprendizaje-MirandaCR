## Tuples

- Tuples are expressed as comma-separated elements within parentheses.
- These are values inside the parentheses.
- In Python, there are different types: strings, integer, float. They can all be contained in a tuple, but the type of the variable is tuple.

```python
tuple1 = ('disco',10,1.2)
type(tuple1)
```

```python
tuple1[0]
```

```python
tuple1[-2]
```

```python
# concatenate
tuple1 + ("hard rock",10)
```

```python
# slicing
tuple1[-3:]
```

```python
len(tuple1 + ("hard rock",10))
```

```python
# Inmutable
Ratings = (10,9,6,5,10,8,9,6,2)
Ratings1 = Ratings

Ratings[2] = 4
```

```python
RatingsSorted = sorted(Ratings)
RatingsSorted 
```

```python
# Tuple nesting
NT = (1,2,("pop","rock"),(3,4),("disco",(1,2)))
NT
```

```python
# Indexes
NT[2][1]
```

## Sets

Sets are a type of collection. This means that like lists and tuples, you
can input different python types. Unlike lists and tuples they are unordered.
They are unique

```python
set1 = {"pop","rock","soul","rock","disco"}

set1
```

```python
l1 = ["A","A","B","B","C"]
l1_seted = set(l1)
l1_seted
```

```python
# Set Operations
A = {"Thriller","Back in Black", "AC/DC"}

# Include element
A.add("NSYNC")
A
```

```python
# Remove element
A.remove("NSYNC")
A
```

```python
# Element in Set
"AC/DC" in A
```

```python
"Who" in A
```

```python
# Mathematical set operations

album_set_1 = {"Thriller","Back in Black", "AC/DC"}
album_set_2 = {"The Dark Side of the Moon","Back in Black", "AC/DC"}


# Intersection
album_set_3 = album_set_1 & album_set_2
album_set_3
```

```python
# Union
album_set_1.union(album_set_2)
```

```python
# Subset
album_set_3.issubset(album_set_1)
```

## Dictionaries

- Denoted with curly brakets {}
- Keys are inmutable and unique
- Values can be inmutable, mutable and duplicates
- Each key and value pair is separed by a comma

```python
DICT = {"Thriller":1982,
        "Back in Black":1980
        }

# Include element
DICT["Rumors"] = 1977

# Delete element
# del(DICT["Rumors"])

DICT["Thriller"]
```

```python
# element in dict
"Rumors" in DICT
```

```python
# Keys
DICT.keys()
```

```python
# Values
DICT.values()
```
