## Branching

Is used to run different statement for a different input


````python
age = 19

# IF
if(age > 18):
  print("you can enter")
print("move on")
````

````python
# IF ELSE

age = 17

if(age > 18):
  print("you can enter")
else:
  print("go see Meat Loaf")
print("move on")
````

````python
# ELIF

age = 18

if(age > 18):
  print("you can enter")
elif(age == 18):
  print("go see Pink Floyd")
else:
  print("go see Meat Loaf")
print("move on")
````

## Class

````python
#Define atributes to initialize each instance of class

class Circle(object):
  def __init__(self,radius,color):
    self.radius=radius;
    self.color=color;

RedCircle = Circle(10,"red")
RedCircle.radius
````

## Method
Functions that interact and change

````python
class Circle(object):
  def __init__(self,radius,color):
    self.radius=radius;
    self.color=color;

  def add_radius(self,r):
    self.radius = self.radius + r


RedCircle = Circle(2,"red")
RedCircle.add_radius(8)
RedCircle.radius
````
